```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
%matplotlib inline


from sklearn import datasets, linear_model
dt = [[13,16,21],[69032,69035,69044]]
tab = np.array(dt)


data_X_test = tab[0].reshape(-1, 1)
data_y_test = tab[1].reshape(-1, 1)
regr = linear_model.LinearRegression()
regr.fit(data_X_test, data_y_test)


pyplot.title('Heure/compt')
pyplot.xlabel('Heure')
pyplot.ylabel('Tarif')
pyplot.scatter(data_X_test, data_y_test,  color='blue')
pyplot.plot(data_X_test, regr.predict(data_X_test), color='red',
         linewidth=3)


pyplot.show()
```
